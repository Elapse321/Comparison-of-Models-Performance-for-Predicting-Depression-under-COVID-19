```python
import copy
def employmentfunct(X):
    Y = copy.deepcopy(X["employment_status_1"])
    for i in range(len(X["employment_status_1"])):
        if (X["employment_status_1"][i]=="Yes" ):
            Y[i] = "Full time employment"
        elif (X["employment_status_2"][i]=="Yes"):
            Y[i] = "Part time employment"
        elif (X["employment_status_3"][i]=="Yes"):
            Y[i] = "Full time student"
        elif (X["employment_status_4"][i]=="Yes"):
            Y[i] = "Retired"
        elif (X["employment_status_5"][i]=="Yes"):
            Y[i] = "Unemployed"
        elif (X["employment_status_6"][i]=="Yes"):
            Y[i] = "Not working"
        else:
            Y[i] = "Other"
    return Y
        

employmentchina = employmentfunct(chinaCOVID19)
employmentdenmark = employmentfunct(denmarkCOVID19)
employmentfinland = employmentfunct(finlandCOVID19)
employmentnorway = employmentfunct(norwayCOVID19)
employmentsweden = employmentfunct(swedenCOVID19)
employmentvietnam = employmentfunct(vietnamCOVID19)

    
```
